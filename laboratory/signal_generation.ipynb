```python
import numpy as line
import matplotlib.pyplot as grph

def unit_sample(n):
    return line.where(n == 0, 3, 0)

x = line.arange(-10, 11)

y = unit_sample(x)

grph.stem(x, y)
grph.title('UNIT SAMPLE SEQUENCE')
grph.xlabel('x')
grph.ylabel('f(t)')
grph.grid()
grph.xlim([-10, 10])
grph.ylim([-1, 5])
grph.axhline(0, color='red', lw=0.5)
grph.axvline(0, color='blue', lw=0.5)
grph.show()
```

```python
import numpy as line
import matplotlib.pyplot as grph

def unit_step(n):
    return line.where(n >= 0, 3, 0)

x = line.arange(-10, 11)

y = unit_step(x)

grph.step(x, y, where='post', label='Unit Step Signal', linewidth=2)
grph.title('UNIT STEP SIGNAL')
grph.xlabel('x')
grph.ylabel('f(t)')
grph.grid()
grph.xlim([-10, 10])
grph.ylim([-1, 5])
grph.axhline(0, color='red', lw=0.5)
grph.axvline(0, color='blue', lw=0.5)
grph.legend()
grph.show()
```

```python
import numpy as line
import matplotlib.pyplot as grph

def unit_ramp(n):
    return line.where(n >= 0, n, 0)

x = line.arange(-10, 11)

y = unit_ramp(x)

grph.plot(x, y, label='Unit Ramp Signal', linewidth=3)
grph.title('UNIT RAMP SIGNAL')
grph.xlabel('x')
grph.ylabel('f(t)')
grph.grid()
grph.xlim([-10, 10])
grph.ylim([-1, 10])
grph.axhline(0, color='red', lw=0.5)
grph.axvline(0, color='blue', lw=0.5)
grph.legend()
grph.show()
```

```python
import numpy as line
import matplotlib.pyplot as grph

A = 10
k = 1
x = line.linspace(-10, 10, 1000)

def exponential_signal(x, A, k):
    return A * line.exp(k * x)

y = exponential_signal(x, A, k)

grph.plot(x, y, label='Exponential Signal', linewidth=2)
grph.title('EXPONENTIAL SIGNAL')
grph.xlabel('x')
grph.ylabel('f(t)')
grph.grid()
grph.xlim([-10, 10])
grph.ylim([0, line.max(x) * 1.1])
grph.axhline(0, color='red', lw=0.5)
grph.axvline(0, color='blue', lw=0.5)
grph.legend()
grph.show()
```

```python
import numpy as line
import matplotlib.pyplot as grph

A = 3
T = 1.5
x = line.linspace(0, 5, 500)

def square_wave(x, A, T):
    return A * line.sign(line.sin(2 * line.pi * x / T))

y = square_wave(x, A, T)

grph.plot(x, y, label='Square Wave Signal', linewidth=2)
grph.title('SQUARE WAVE SIGNAL')
grph.xlabel('x')
grph.ylabel('f(t)')
grph.grid()
grph.xlim([0, 5])
grph.ylim([-5, 5])
grph.axhline(0, color='red', lw=0.5)
grph.axvline(0, color='blue', lw=0.5)
grph.legend()
grph.show()
```

```python
import numpy as line
import matplotlib.pyplot as grph

A = 1.5
T = 3
x = line.linspace(0, 10, 500)

def triangular_wave(x, A, T):
    return A * (2 / T) * line.abs((x % T) - T / 2) - A / 2

y = triangular_wave(x, A, T)

grph.plot(x, y, label='Triangular Signal', linewidth=2)
grph.title('TRIANGULAR SIGNAL')
grph.xlabel('x')
grph.ylabel('f(t)')
grph.grid()
grph.xlim([0, 10])
grph.ylim([-3, 3])
grph.axhline(0, color='red', lw=0.5)
grph.axvline(0, color='blue', lw=0.5)
grph.legend()
grph.show()
```
